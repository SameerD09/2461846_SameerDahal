Task 6:

• Create a new column HDI_Category that classifies each country into one of the four official
Human Development Index groups. The classification should be based on the HDI value for the
year 2022. Use the following categories and thresholds defined by the United Nations Development

Programme (UNDP):

HDI Category | HDI Range (hdi)

Low          | < 0.550

Medium       | 0.550 – 0.699

High         | 0.700 – 0.799

Very         | ≥ 0.800

After creating this new column:

• verify that all countries are classified correctly,

• ensure the updated dataframe includes the new category column.

• Save the final dataframe as HDI_category_added.csv and include this file in your final submission.


```python
# 1) Create HDI_Category based on UNDP thresholds
#    Low:    < 0.550
#    Medium: 0.550 - 0.699
#    High:   0.700 - 0.799
#    Very:   >= 0.800
def assign_hdi_category(hdi_value):
    """Return UNDP HDI category based on 2022 HDI value."""
    if pd.isna(hdi_value):
        return np.nan
    elif hdi_value < 0.550:
        return "Low"
    elif 0.550 <= hdi_value <= 0.699:
        return "Medium"
    elif 0.700 <= hdi_value <= 0.799:
        return "High"
    else:
        return "Very"

# Apply the categorization rule to create the new column
hdi_2022_df["HDI_Category"] = hdi_2022_df["hdi"].apply(assign_hdi_category)

# 2) Verify classification and ensure column exists
print("HDI_Category column added. Category counts:")
display(hdi_2022_df["HDI_Category"].value_counts(dropna=False))

print("\nSample check (country, hdi, HDI_Category):")
display(hdi_2022_df[["country", "hdi", "HDI_Category"]].head(10))

# 3) Save the final dataframe as HDI_category_added.csv
output_path = "/content/drive/MyDrive/HDI_Assignment_1/HDI_category_added.csv"
hdi_2022_df.to_csv(output_path, index=False)

print(f"\nSaved file successfully to: {output_path}")
```

Task 7:

• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 × IQR rule.

• Create scatter plots highlighting outliers in a different color.

```python
# Load the required data
south_asia_df = pd.read_csv('/content/drive/MyDrive/HDI_Assignment_1/HDI_SouthAsia_2020_2022.csv')
middle_east_df = pd.read_csv('/content/drive/MyDrive/HDI_Assignment_1/HDI_MiddleEast_2020_2022.csv')

# Function to calculate outliers using 1.5 * IQR rule
def detect_outliers(df, column):
    # Calculate the first (Q1) and third (Q3) quartiles for the data
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    # Compute the IQR
    IQR = Q3 - Q1

    # Calculate the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    return outliers, lower_bound, upper_bound

# 1. Detect outliers for HDI and GNI per Capita for South Asia
south_asia_hdi_outliers, south_asia_hdi_lower, south_asia_hdi_upper = detect_outliers(south_asia_df, "hdi")
south_asia_gni_outliers, south_asia_gni_lower, south_asia_gni_upper = detect_outliers(south_asia_df, "gross_inc_percap")

# 2. Detect outliers for HDI and GNI per Capita for Middle East
middle_east_hdi_outliers, middle_east_hdi_lower, middle_east_hdi_upper = detect_outliers(middle_east_df, "hdi")
middle_east_gni_outliers, middle_east_gni_lower, middle_east_gni_upper = detect_outliers(middle_east_df, "gross_inc_percap")

# 3. Create scatter plots highlighting outliers in different colors

# Scatter plot for HDI vs GNI for South Asia
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gross_inc_percap', y='hdi', data=south_asia_df, label='Normal Data', color='blue')
sns.scatterplot(x=south_asia_hdi_outliers['gross_inc_percap'], y=south_asia_hdi_outliers['hdi'], label='Outliers', color='red')
plt.title('South Asia: HDI vs GNI with Outliers')
plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.legend()
plt.tight_layout()
plt.show()

# Scatter plot for HDI vs GNI for Middle East
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gross_inc_percap', y='hdi', data=middle_east_df, label='Normal Data', color='blue')
sns.scatterplot(x=middle_east_hdi_outliers['gross_inc_percap'], y=middle_east_hdi_outliers['hdi'], label='Outliers', color='red')
plt.title('Middle East: HDI vs GNI with Outliers')
plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.legend()
plt.tight_layout()
plt.show()

# 4. Display the detected outliers
print("South Asia HDI Outliers:")
print(south_asia_hdi_outliers[['country', 'hdi', 'gross_inc_percap']])

print("\nMiddle East HDI Outliers:")
print(middle_east_hdi_outliers[['country', 'hdi', 'gross_inc_percap']])

print("\nSouth Asia GNI Outliers:")
print(south_asia_gni_outliers[['country', 'gross_inc_percap', 'hdi']])

print("\nMiddle East GNI Outliers:")
print(middle_east_gni_outliers[['country', 'gross_inc_percap', 'hdi']])
```
